```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.ensemble import AdaBoostClassifier

# Split data into training and testing sets, ensuring stratification
# This ensures the proportion of 'Pass' and 'Fail' is similar in train and test sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=.2, random_state=42, stratify=Y)

# Identify categorical and numerical columns for encoding
categorical_cols = X_train.select_dtypes(include='object').columns
numerical_cols = X_train.select_dtypes(exclude='object').columns

# Initialize OneHotEncoder
ohe = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

# Fit and transform on X_train categorical columns
X_train_categorical_encoded = ohe.fit_transform(X_train[categorical_cols])
X_train_categorical_df = pd.DataFrame(X_train_categorical_encoded, columns=ohe.get_feature_names_out(categorical_cols), index=X_train.index)

# Transform X_test categorical columns (using the encoder fitted on X_train)
X_test_categorical_encoded = ohe.transform(X_test[categorical_cols])
X_test_categorical_df = pd.DataFrame(X_test_categorical_encoded, columns=ohe.get_feature_names_out(categorical_cols), index=X_test.index)

# Combine numerical and one-hot encoded categorical columns for X_train
X_train_processed = pd.concat([X_train[numerical_cols], X_train_categorical_df], axis=1)

# Combine numerical and one-hot encoded categorical columns for X_test
X_test_processed = pd.concat([X_test[numerical_cols], X_test_categorical_df], axis=1)

# Initialize and train the Decision Tree Classifier with processed data
dtc_classifier = DecisionTreeClassifier(random_state=42)
dtc_classifier.fit(X_train_processed, Y_train)

print("Decision Tree Classifier trained successfully with one-hot encoded features and new target.")

# Predict and evaluate the Decision Tree Classifier
Y_pred_dtc = dtc_classifier.predict(X_test_processed)
accuracy_dtc = accuracy_score(Y_test, Y_pred_dtc)
print(f"Decision Tree Classifier Accuracy: {accuracy_dtc}")

# Initialize and train the AdaBoost Classifier with processed data
adaboost_classifier = AdaBoostClassifier(random_state=42)
adaboost_classifier.fit(X_train_processed, Y_train)

print("AdaBoost Classifier trained successfully with one-hot encoded features and new target.")
```

```python
dt_rf = RandomForestClassifier(random_state=42)
dt_rf.fit(X_train_processed, Y_train)
```

```python
dt_rf_pred = dt_rf.predict(X_test_processed)
ada_rf_pred = adaboost_classifier.predict(X_test_processed)
dtc_rf_pred = dtc_classifier.predict(X_test_processed)
dt_rf_acc = accuracy_score(Y_test, dt_rf_pred)
ada_rf_acc = accuracy_score(Y_test, ada_rf_pred)
dtc_rf_acc = accuracy_score(Y_test, dtc_rf_pred)


print(dt_rf_acc)
print(ada_rf_acc)
print(dtc_rf_acc)
```
